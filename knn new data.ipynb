# imports

```python
import numpy as np 
import pandas as pd 
from sklearn.cluster import KMeans 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from sklearn.preprocessing import normalize
from sklearn.metrics import pairwise_distances
from nltk.stem.porter import *
stemmer=SnowballStemmer("english")
import os, sys, email,re
import nltk
import gensim
import string
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')

data=pd.read_csv("abcnews-date-text.csv",nrows=5000)
```

```python
data=data["headline_text"]
```

# cleaning data

```python

def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))


def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) >= 3:
            result.append(lemmatize_stemming(token))
    return result


```

```python
data=data.map(preprocess)
```

```python
for i in range(len(data)):
    data[i]=" ".join(data[i])
```

```python
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vectorizor = TfidfVectorizer(stop_words = 'english',#tokenizer = tokenize_and_stem,
                             max_features = 30000)
tf_idf = tf_idf_vectorizor.fit_transform(data)
tf_idf_norm = normalize(tf_idf)
tf_idf_array = tf_idf_norm.toarray()
```

```python
data3=pd.DataFrame(tf_idf_array, columns=tf_idf_vectorizor.get_feature_names())
```

```python
data3
```

# k mean clustering class

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=10,init='k-means++',max_iter=10,n_init=10)
km.fit(data3)
```

```python
order_centroids =km.cluster_centers_.argsort()[:, ::-1]
terms = tf_idf_vectorizor.get_feature_names()
for i in range(15):
    print("Cluster %d:" % i)
    for j in order_centroids[i, :10]: #print out 10 feature terms of each cluster
        print (' %s' % terms[j])
    print('------------')
```

```python

from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=3, n_iter=7, random_state=42)
scatter_plot_points = svd.fit_transform(data3)
scatter_plot_points=pd.DataFrame(scatter_plot_points,columns=["x","y","z"])

import plotly.express as px
fig = px.scatter_3d(scatter_plot_points,x="x",y="y",z="z",color=km.labels_)
fig.show()
```
